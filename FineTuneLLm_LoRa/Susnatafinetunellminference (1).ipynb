```python
from unsloth import FastLanguageModel

# 1. Loading  model
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "/content/final_model/final_model",
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)
```

```python
FastLanguageModel.for_inference(model)

# Example prompt
prompt = "Explain North atlantic treaty organisation in simple words with an example."

# Tokenize input
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)



# Generate response
outputs = model.generate(
    **inputs, ## kwargs unpacking as input are 2d dict containing token ids and attentions masks
    max_new_tokens = 256,    # control answer length
    temperature = 0.7,       # creativity (lower = logical, higher = creative)
    top_p = 0.9,             # nucleus sampling
    do_sample = True,        # sampling instead of greedy decoding
)

# Decode output
print(tokenizer.decode(outputs[0], skip_special_tokens=True))  #batch decode helps me to convert ids back into natural language so that i can see
```

```python
from unsloth.chat_templates import get_chat_template
from unsloth import FastLanguageModel

#  System prompt (defines model personality)
sys_prompt = """You are a reflective assistant engaging in thorough,
step-by-step reasoning, and answering clearly."""

#  User message
user_message = "If it's raining, why is the sky grey?"

#  Loading chat template for my tokenizer
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "llama-3.1",   # keep consistent with your base
)

#  Switching to inference mode
FastLanguageModel.for_inference(model)

#  Wrapping messages in chat format
messages = [
    {"role": "system", "content": sys_prompt},
    {"role": "user", "content": user_message},
]

#  Applying template
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,
    add_generation_prompt = True,  # must add for generation
    return_tensors = "pt",
).to(model.device)

#  Generating
outputs = model.generate(
    input_ids = inputs,
    max_new_tokens = 512,
    use_cache = True,
    temperature = 0.7,  # balance reasoning + creativity
    min_p = 0.1, #Hf style relative to highest probability
    #top_p=0.9
    do_sample = True,
)

#  Decoding
response = tokenizer.batch_decode(outputs, skip_special_tokens=True)   #batch decode helps me to convert ids back into natural language so that i can see
print(response[0])
```
